```python
DEVELOPER_KEY = "AIzaSyD42avPBMPzSPo31ySL_uoG746XkTCd-1w"
YOUTUBE_API_SERVICE_NAME = "youtube"
YOUTUBE_API_VERSION = "v3"
youtube = build(YOUTUBE_API_SERVICE_NAME, YOUTUBE_API_VERSION, developerKey=DEVELOPER_KEY)
```

```python
playlist_videos = []

res = youtube.playlistItems().list(
  part='snippet', playlistId = playlistID, maxResults = 30).execute()
```

```python
playlist_videos = res['items']
display(playlist_videos)
```

```python
playlist_videos[0]['kind']
```

```python
videos_ids = list(map(lambda x: x['snippet']['resourceId']['videoId'], playlist_videos))
```

```python
stats = []
for video_id in videos_ids:
  response = youtube.videos().list(part = 'statistics', id = video_id).execute()
  stats += response['items']
```

```python
videos_title = list(map(lambda x: x['snippet']['title'], playlist_videos))
#url_thumbnails = list(map(lambda x: x['snippet']['thumbnails']['high']['url'], playlist_videos))
published_date = list(map(lambda x: str(x['snippet']['publishedAt']), playlist_videos))
video_id = list(map(lambda x: x['snippet']['resourceId']['videoId'], playlist_videos))
print(videos_title)
```

```python
likes = list(map(lambda x: int(x['statistics']['likeCount']), stats))
favorite = list(map(lambda x: int(x['statistics']['favoriteCount']), stats))
views = list(map(lambda x: int(x['statistics']['viewCount']), stats))
comment = list(map(lambda x: int(x['statistics']['commentCount']), stats))
print(likes)
```

```python
from datetime import datetime
import pandas as pd
```

```python
playlist_df = pd.DataFrame(
    {'title': videos_title,
     'published_date': published_date,
     'video_id': video_id,
     'likes': likes,
     'views': views,
     'comment': comment})

display(playlist_df)
```

```python
playlist_df.to_csv("youtube_playlist.csv", index = False)
```
