```python
import json
import requests
import pandas as pd
import csv
```

```python
nomes = ['leonardo', 'pedro', 'lucas', 'matheus', 'davi', 'francis', 'francisco', 'sebastião', 'enzo', 'otávio']
lista_respostas = []
url = ('https://servicodados.ibge.gov.br/api/v2/censos/nomes/')
```

```python
for nome in nomes:
  requisicao = requests.get(url + nome)  
  if requisicao.status_code == 200:
    lista_respostas.append(requisicao.json())
  else: 
    print('Error: ' + str(requisicao.status_code))      

# Quando usar o json_normalize???
dados_IBGE = pd.json_normalize(lista_respostas)
#print(dados_IBGE)
```

```python
lista_respostas_final = []
for indice in range(0,len(nomes)):
  lista_respostas_final += lista_respostas[indice]
```

```python
nomes_df = pd.DataFrame(data = lista_respostas_final[0])
for i in range(1,len(nomes)):
  nomes_df = nomes_df.append(pd.DataFrame(data = lista_respostas_final[i]))

nomes_df.pop('sexo')
nomes_df.pop('res')
nomes_df.pop('localidade')
nomes_df.head()
```

```python
# lista_nomes = list(map(lambda x: x['nome'], lista_respostas_final))
# lista_localidade = list(map(lambda x: x['localidade'], lista_respostas_final))
lista_res = list(map(lambda x: x['res'], lista_respostas_final))

lista_periodo = list()
lista_frequencia = list()
for i in range(0,10):
  for j in range(0,9):
    lista_periodo.append(lista_res[i][j]['periodo'])
    lista_frequencia.append(lista_res[i][j]['frequencia'])

nomes_df['periodo'] = lista_periodo
nomes_df['frequencia'] = lista_frequencia
```

```python
nomes_df.to_csv("nomes_ibge.csv", index = False)
```
